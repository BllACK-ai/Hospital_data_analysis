```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import re
```

```python
ConCurrent = "Clean_Name.csv"
Data = pd.read_csv(ConCurrent)
Data
```

```python
Data["AGE"] = Data["AGE"].str.strip()
Data["AGE"] = Data["AGE"].replace(["N/A", "n/a", "-3", "NA", "null", "---"], np.nan)
Data["AGE"] = Data["AGE"].replace("38.0", "38")
Data["AGE"] = Data["AGE"].replace("36.0", "36")
```

```python
Data["AGE"] = Data["AGE"].astype("Int64") #the Int64 allows for no error when converting to int if column has NaN values

# Data.to_csv("Clean_Name.csv", index=False)
```

```python
# Data["AGE"] = Data["AGE"].replace(np.nan, 46) this works logically but
Data["AGE"] = Data["AGE"].fillna(46) #but this is designed to handle missing values
```

```python
Data.to_csv("Clean_AGE.csv", index =False)
```
